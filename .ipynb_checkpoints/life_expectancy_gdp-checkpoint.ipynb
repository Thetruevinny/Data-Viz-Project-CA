# Code Academy Data Visualization Project

### Importing Relevant Libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
```

### Loading Data

```python
df_alpha = pd.read_csv('all_data.csv')
print(df_alpha.head())
```

### Inspecting and Cleaning the Data

```python
print(df_alpha.info())
```

### Analysis

#### Has Life Expectancy Increased for All Nations Over the Years?

```python
fig = plt.figure()
ax = plt.subplot(1,1,1)

for i in df_alpha.Country.value_counts().index:
    sub_df = df_alpha[df_alpha['Country'] == i]
    plt.plot(sub_df.Year, sub_df['Life expectancy at birth (years)'], label = i)

plt.title('Life Expectancy over Time for Six Different Nations')
plt.axis([1999, 2016, 40, 85])
plt.xlabel('Years')
plt.ylabel('Life Expectancy at birth (years)')
plt.legend(loc = 'center left', bbox_to_anchor = (1, 0.5)) 
plt.show()
plt.clf()
```

As can be seen from the graph above the lif expectancy for all 6 countries has roughly been increasing from 2000 to 2016, with Zimbabwe having seen the largest increase. In addition to this Germany's life expectancy has always been the highest. However Chile is approaching Germany's life expectancy.

#### Has GDP Increased For All Six Nations During these Years?

```python
df_alpha.GDP = df_alpha.GDP.apply(lambda x : np.log(x))

fig = plt.figure()
ax = plt.subplot(1,1,1)

for i in df_alpha.Country.value_counts().index:
    sub_df = df_alpha[df_alpha['Country'] == i]
    plt.plot(sub_df.Year, sub_df['GDP'], label = i)

plt.title('GDP over Time for Six Different Nations')
plt.axis([1999, 2016, 22, 32])
plt.xlabel('Years')
plt.ylabel('log(GDP)')
plt.legend(loc = 'center left', bbox_to_anchor = (1, 0.5)) 
plt.show()
plt.clf()
```

Over these years, the GDP has generally increased for all nations. However during this period we have had some countries see years where the GDP has declined in particular Zimbabwe and Mexico. The US remains the largest country in terms of GDP over this period. However, China is rapdily approaching the US in terms of GDP by 2015 as it is the country with the largest growth over this time period.

#### Is there a correlation between Life Expectancy and GDP?

```python
sns.scatterplot(data = df_alpha, x = 'Life expectancy at birth (years)', y  = 'GDP', hue = 'Country')
plt.ylabel('log(GDP)')
plt.title('How GDP varies with Life Expectancy')
plt.show()
```

From the graph it seems that we may not have a large eneough sample to truly give an interpretation of the link of GDP and life expectancy. However there are early signs that lower life expectancy leads to lower GDPs. In the next lines of code, I will print out the correlation co-efficients for these two variables.

```python
corr, p = pearsonr(df_alpha['Life expectancy at birth (years)'], df_alpha.GDP)
print(corr)
```

The correlation co-efficient seems to suggest a strong positive relationship between the two variables and coincides with what we observed from the graph above.

#### Average Life Expectancy in Each Country Over this Period?

```python
LE_dict = {}

for i in df_alpha.Country.value_counts().index:
    sub_df = df_alpha[df_alpha['Country'] == i]
    avg_LE = np.mean(sub_df['Life expectancy at birth (years)'])
    LE_dict[str(i)] = avg_LE

print(LE_dict)
```

These values give a reasonable idea for the life expectancy across the period of 2000 - 2015 for all countries in the list above, apart from Zimbabwe. This is because the other countries in the list did not see non-linear changes in life expectancy during this period (as can be seen from the first graph in the document). Zimbabwe's life expectancy over this period may not be best represented by an average as it saw an initial decline in the beginning followed by rapid increase. This may have skewed the mean to be higher than expected for Zimbabwe during this period, as Zimbabwe's life expectancy is below 50 for more than half this time period.
